```python
#SymPy’s integrals f(x) = x^10 - 55x^9 + 1320x^8 - 18150x^7 + 157773x^6 - 902055x^5 + 3416930x^4 - 8409500x^3 + 12753576x^2 - 10628640x + 3628800
```

```python
import sympy
result = {}  
x = sympy.Symbol("x")
i = sympy.integrate(x**10 - 55*x**9 + 1320*x**8 - 18150*x**7 + 157773*x**6 - 902055*x**5 + 3416930*x**4 - 8409500*x**3 + 12753576*x**2 - 10628640*x + 3628800)
print(i)
result["analytical"] = float(i.subs(x, 10) - i.subs(x, 0))
print("Analytical result: {}".format(result["analytical"]))
```

```python
#Integrals with Monte Carlo method f(x) = x^10 - 55x^9 + 1320x^8 - 18150x^7 + 157773x^6 - 902055x^5 + 3416930x^4 - 8409500x^3 + 12753576x^2 - 10628640x + 3628800
```

```python
import numpy
N = 1_000_000
accum = 0
for i in range(N):
    x = numpy.random.uniform(0, 10)
    accum += x**10 - 55*x**9 + 1320*x**8 - 18150*x**7 + 157773*x**6 - 902055*x**5 + 3416930*x**4 - 8409500*x**3 + 12753576*x**2 - 10628640*x + 3628800
volume = 10 - 0
result["MC"] = volume * accum / float(N)
print("Standard Monte Carlo result: {}".format(result["MC"]))
```

```python
#Integrals with Monte Carlo method Example f(x) = x^3 * sin(x)^3 * exp(-x)
```

```python
import numpy
from sympy import sin, exp
N = 100_000
accum = 0
result =[]
for i in range(N):
    x = numpy.random.uniform(0, 10)
    accum += x**3 * sin(x)**3 * exp(-x)
volume = 10 - 0
result = volume * accum / float(N)
print("Standard Monte Carlo result: {}".format(result))
```
