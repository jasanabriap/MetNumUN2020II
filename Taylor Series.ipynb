```python
#Create a function to estimate the value of cos(x) using a Taylor Series
```

```python
import math
def func_cos(x, n):
    cos_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i)
        denom = math.factorial(2*i)
        cos_approx += ( coef ) * ( (num)/(denom) )
    return cos_approx
angle_rad = (math.radians(45)) #convert 45 degrees into radians
out = func_cos(angle_rad,5)
print(out)
```

```python
print(math.cos(angle_rad))
```

```python
#Build a plot to compare the Taylor Series approximation to Python's cos() function
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

angles = np.arange(-2*np.pi,2*np.pi,0.1)
p_cos = np.cos(angles) #cos function
t_cos = [func_cos(angle,3) for angle in angles] #Taylor Series - 3 terms

fig, ax = plt.subplots()
ax.plot(angles,p_cos)
ax.plot(angles,t_cos)
ax.set_ylim([-2,5])
ax.legend(['cos() function','Taylor Series - 3 terms'])

plt.show()
```

```python
#Use a for loop to see how much better adding additional terms to our Taylor Series approximation compares to Python's cos() function
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

angles = np.arange(-2*np.pi,2*np.pi,0.1)
p_cos = np.cos(angles)

fig, ax = plt.subplots()
ax.plot(angles,p_cos)

for i in range(1,6): #lines for 1 - 6 terms in the Taylor Series
    t_cos = [func_cos(angle,i) for angle in angles] 
    ax.plot(angles,t_cos)

ax.set_ylim([-7,4]) #dom

legend_lst = ['cos() function']
for i in range(1,6):
    legend_lst.append(f'Taylor Series - {i} terms')
ax.legend(legend_lst, loc=0)

plt.show()
```

```python
#Taylor Series Expansion and its special case Maclaurin Series Expansion order(n) = 4
```

```python
from sympy import series, Symbol
from sympy.functions import sin, cos, exp
from sympy.plotting import plot
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = 13,10
plt.rcParams['lines.linewidth'] = 2

x = Symbol('x')

def taylor(function, x0, n): 
    #function is the function to approximate
    #x0 is the point in to approximate
    #n is the order of approximation
    return function.series(x,x0,n).removeO()
```

```python
#Expansion order(n) = 4
print('sin(x) =', taylor(sin(x), 0, 4))
print('cos(x) =', taylor(cos(x), 0, 4))
print('e(x) =', taylor(exp(x), 0, 4))
```

```python
#Evaluate these functions x = 1
print('sin(1) =', taylor(sin(x), 0, 4).subs(x,1))
print('cos(1) =', taylor(cos(x), 0, 4).subs(x,1))
print('e(1) =', taylor(exp(x), 0, 4).subs(x,1))
```

```python
import math

x = 1
epsilon = 10**(-4)

for k in range(1000):
  if ((math.log(1+x) * x**(k+1)) / math.factorial(k+1)  < epsilon):
    break
print(k)

f_x = 0
for i in range(k+1):
    i+=1
    print(i,' ',f_x, ' delta ',((-1)**(i-1))*(x**i)/i)
    f_x += ((-1)**(i-1))*x**i/i

print('ln(1+1) = ',math.log(1+1))
```

```python
import math

x = 1
epsilon = 10**(-16)

for k in range(1000):

  if (  ( (x * x**(k+1)) / math.factorial(k+1) ) < epsilon):

    break
print(k)

f_x = 0
for i in range(k+1): 
    i+=1
    delta=((-1)**(i-1))*(x**i)/i
    if(delta<0):
     delta=delta*-1 
    print(i-1,' ',f_x, ' delta ',delta)
    f_x += ((-1)**(i-1))*x**i/i

print('ln(1+1) = ',math.log(1+1))
```

```python
import math

x = 1
epsilon = 10**(-4)
delta = 1
f_x = 0
i = 1
while  (epsilon <= delta ):
  delta = ((-1)**(i-1))*x**i/i
  f_x += delta 
  if(delta<0):
    delta=delta*-1
  i = i+1
print(i-1) 
print('f_x = ',f_x)
print('ln(1+1) = ',math.log(1+1))
```
